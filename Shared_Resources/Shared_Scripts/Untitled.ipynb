```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for UBS stock over the last 20 years
ubs_stock = yf.download('NVIDIA', start='1990-01-01', end='2023-01-01')

# Plotting the closing prices
plt.figure(figsize=(14, 7))
plt.plot(ubs_stock['Close'], label='UBS Closing Price')
plt.title('UBS Stock Price Over the Last 20 Years')
plt.xlabel('Year')
plt.ylabel('Price in USD')
plt.legend()
plt.grid(True)
plt.show()
```
